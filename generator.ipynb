```python
np.random.seed(4)
```

```python
T = 200
GC1 = np.array([[1,0,1],[1,1,1],[0,0,0]])
GC2 = np.array([[1,0,1],[0,1,0],[0,0,0]])
GC_l = np.zeros([T,3,3])
GC_l[0] = GC1
alpha = 0.5
alpha_l = np.zeros(T)
X = np.zeros([T,3])
X[:,2] = np.random.randint(-1,2,size=T)
alpha_l[T//4:3*T//4] = alpha
for i in range(1,T):
    X[i,0] = X[i-1,0] + X[i-1,2]
    X[i,1] = alpha_l[i] * X[i-1,2] * X[i-1,0]
    if alpha_l[i] == 0:
        GC_l[i] = GC2
    else:
        GC_l[i] = GC1
```

```python
plt.plot(X)
```
